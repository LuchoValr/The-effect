#Variables instrumentales con regresion a dos partes

```python
!pip install causaldata
!pip install linearmodels
import pandas as pd
from linearmodels.iv import IV2SLS
from causaldata import social_insure
```

```python
d = social_insure.load_pandas (). data
```

```python
# Create an control -variable DataFrame
# including dummies for village
controls = pd.concat([d[['male','age','agpop','ricearea_2010',
                         'literacy','intensive','risk_averse','disaster_prob']], pd.get_dummies(d[['village']])],axis = 1)
```

```python
# Create model and fit separately
# since we want to cluster , and will use
# m.notnull to see which observations to drop
m = IV2SLS(d['takeup_survey'],
           controls , d['pre_takeup_rate'], d['default'])
```

```python
second_stage = m.fit(cov_type = 'clustered', clusters = d['address'][m.notnull])
```

```python
# If we want the first stage we must do it ourselves!
first_stage = IV2SLS(d['pre_takeup_rate'], pd.concat([controls ,d['default']],axis = 1),None ,
                     None).fit(cov_type = 'clustered', clusters = d['address'][m.notnull])
```
